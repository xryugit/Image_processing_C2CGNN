```python
def iq_getcameraqe(desc, wavel, *varargs):
    justpaths = False
    if varargs:
        for arg in varargs:
            if arg == "justpaths":
                justpaths = True
    desc = desc[0]
    iqp = iq_getspectralpar()
    wavel = np.array(list(iqp['WaveLength'].values())[0],dtype='float64')
    cdb = r"/Users/xinranyu/Desktop/1/Code/Image Analysis/CameraSpectra.xlsx"
    if justpaths:
        return cdb

    cams = {
        'Zyla5': 'zyla\s{0,3}(vsc.+(00829|00369)|5)',
        'Prime95B': 'prime',
        'Kinetix': 'kinetix'
    }

    for key, pattern in cams.items():
        if re.search(pattern, desc, re.IGNORECASE):
            camname = key
            break
    else:
        raise ValueError("No camera found for that name")

    db = ReadCamDB(cdb, cams)
    f = interp1d(db['Zyla5'][:, 0].astype(np.float64), db['Zyla5'][:, 1].astype(np.float64), kind='linear', fill_value="extrapolate")
    qe = f(wavel)
    return qe, cdb,db

```
